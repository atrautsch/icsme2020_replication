```python
def lower_bound(bug_matrix, efforts, predictions):
    # bug_matrix = bug_matrix.loc[:,bug_matrix.sum()>0]  # is already taken care of in get_labels
    effort_true = efforts[predictions].sum()
    bugs_found = bug_matrix.sum().eq(bug_matrix[predictions].sum()).sum()
    return effort_true/bugs_found

def upper_bound(bug_matrix, efforts, predictions):
    # bug_matrix = bug_matrix.loc[:,bug_matrix.sum()>0]  # is already taken care of in get_labels
    effort_false = efforts[~predictions].sum()
    bugs_missed = len(bug_matrix.columns)-bug_matrix.sum().eq(bug_matrix[predictions].sum()).sum()
    return effort_false/bugs_missed

def get_cost_labels(df, boundary_date=None):
    labels_bug = {}
    bug_matrix_cols = []
    labels_adhoc = {}
    adhoc_matrix_cols = []
    first_bug = datetime.datetime.now(datetime.timezone.utc)
    first_adhoc = datetime.datetime.now(datetime.timezone.utc)
    for idx, row in df.iterrows():
        labels_bug[idx] = 0
        labels_adhoc[idx] = 0
        for key in df.columns:
            if not key.startswith('adhoc') and len(key.split('__')) > 2 and row[key] > 0:  # we hit the bug matrix with bug for which ROW is inducing
                tripel = key.split('__')  # issue-id__bugfix-revision__bugfix-date
                bugfix_date = pd.to_datetime(tripel[-1])
                bugfix_date = bugfix_date.replace(tzinfo=row['committer_date'].tzinfo)
                if boundary_date and bugfix_date < boundary_date:
                    labels_bug[idx] = 1
                    bug_matrix_cols.append(key)
                    if first_bug > bugfix_date:
                        first_bug = bugfix_date
                if not boundary_date:
                    labels_bug[idx] = 1
                    bug_matrix_cols.append(key)
                    if first_bug > bugfix_date:
                        first_bug = bugfix_date
            if key.startswith('adhoc') and len(key.split('__')) > 2 and row[key] > 0:  # we hit the bug matrix with bug for which ROW is inducing
                tripel = key.split('__')  # issue-id__bugfix-revision__bugfix-date
                bugfix_date = pd.to_datetime(tripel[-1])
                bugfix_date = bugfix_date.replace(tzinfo=row['committer_date'].tzinfo)
                if boundary_date and bugfix_date < boundary_date:
                    labels_adhoc[idx] = 1
                    adhoc_matrix_cols.append(key)
                    if first_adhoc > bugfix_date:
                        first_adhoc = bugfix_date
                if not boundary_date:
                    labels_adhoc[idx] = 1
                    adhoc_matrix_cols.append(key)
                    if first_adhoc > bugfix_date:
                        first_adhoc = bugfix_date
    return list(labels_bug.values()), list(set(bug_matrix_cols)), list(labels_adhoc.values()), list(set(adhoc_matrix_cols)), first_bug, first_adhoc


def get_labels(df, boundary_date=None):
    labels_adhoc = {}
    labels_bug = {}
    for idx, row in df.iterrows():
        labels_adhoc[idx] = 0
        labels_bug[idx] = 0
        for key in df.columns:
            if key.startswith('adhoc') and len(key.split('__')) > 2 and row[key] == 1:  # we hit the bug matrix with bug for which ROW is inducing
                tripel = key.split('__')  # issue-id_bugfix-revision_bugfix-date
                bugfix_date = pd.to_datetime(tripel[-1])
                labels_adhoc[idx] = 1
            if not key.startswith('adhoc') and len(key.split('__')) > 2 and row[key] == 1:
                tripel = key.split('__')  # issue-id_bugfix-revision_bugfix-date
                bugfix_date = pd.to_datetime(tripel[-1])
                labels_bug[idx] = 1
    return list(labels_adhoc.values()), list(labels_bug.values())

def load_project(project_name):
    f = '{}/jit_sn_{}.csv'.format(DATA_PATH, project_name)
    if os.path.exists(f):
        tmp = pd.read_csv(f)
        tmp['project'] = project_name
        tmp['committer_date'] = pd.to_datetime(tmp['committer_date'])
        tmp = tmp.fillna(0)
        return tmp

def score_project(project):
    df = load_project(project)
    if df is None:
        return []

    scores = []

    bug, bug_cols, adhoc, adhoc_cols, first_bug, first_adhoc = get_cost_labels(df)
    
    df['adhoc_label'] = adhoc
    df['bug_label'] = bug
    print(project)
    for name, f in zip(['jit', 'static', 'pmd', 'jit_static_pmd'], [JIT_FEATURES, STATIC_FEATURES, PMD_FEATURES, JIT_FEATURES + STATIC_FEATURES, JIT_FEATURES + STATIC_FEATURES + PMD_FEATURES]):
        for lbl in ['pascarella_commit', 'adhoc_label', 'bug_label']:
            
            X = df[df['la'] > 0]
            y = df[df['la'] > 0][lbl]
            X_train, X_test, y_train, y_test = train_test_split(X, y)
            
            sm = SMOTE(random_state=RANDOM_SEED)
            X_res, y_res = sm.fit_resample(X_train[f].values, X_train[lbl].values)

            X_train_f = X_train[f]
            X_test_f = X_test[f]
            
            scaler = StandardScaler()
            X_train_scaled = scaler.fit_transform(X_res)
            X_test_scaled = scaler.fit_transform(X_test_f)

            lr = LogisticRegression(penalty='elasticnet', l1_ratio=0.5, solver='saga')
            lr.fit(X_train_scaled, y_res)
            lr_y_pred = lr.predict(X_test_scaled)
            
            lr_roc_auc = roc_auc_score(y_test, lr.predict_proba(X_test_scaled)[:,1])
            lr_f1 = f1_score(y_test, lr_y_pred)
    
            # rf
            model = RandomForestClassifier()
            rf_model = model.fit(X_res, y_res)

            rf_y_pred = rf_model.predict(X_test_f)
            proba = rf_model.predict_proba(X_test_f)

            rf_roc_auc = roc_auc_score(y_test, proba[:,1])
            f1 = f1_score(y_test, rf_y_pred)

            lr_ub = lr_lb = rf_ub = rf_lb = 0
            if lbl in ['bug_label', 'adhoc_label']:
                # cost model only
                bug, bug_cols, adhoc, adhoc_cols, _, _ = get_cost_labels(X_test)
                td = X_test.copy()
                td['effort'] = td['la'] + td['ld']
                td['rf_predictions'] = rf_y_pred.astype(bool)
                td['lr_predictions'] = lr_y_pred.astype(bool)
                bug_matrix = td[bug_cols]
                adhoc_matrix = td[adhoc_cols]

            if lbl == 'bug_label':
                rf_ub = upper_bound(bug_matrix, td['effort'], td['rf_predictions'])
                rf_lb = lower_bound(bug_matrix, td['effort'], td['rf_predictions'])
                lr_ub = upper_bound(bug_matrix, td['effort'], td['lr_predictions'])
                lr_lb = lower_bound(bug_matrix, td['effort'], td['lr_predictions'])
            elif lbl == 'adhoc_label':
                rf_ub = upper_bound(adhoc_matrix, td['effort'], td['rf_predictions'])
                rf_lb = lower_bound(adhoc_matrix, td['effort'], td['rf_predictions'])
                lr_ub = upper_bound(adhoc_matrix, td['effort'], td['lr_predictions'])
                lr_lb = lower_bound(adhoc_matrix, td['effort'], td['lr_predictions'])
            
            print('features', name)
            print(lbl, 'positives: {} of {}'.format(np.sum(y), len(y)))
            print('RF AUC-ROC: {}'.format(rf_roc_auc * 100))
            print('RF F1: {}'.format(f1))
            scores.append({'project': project, 'label': lbl, 'metric_set': name, 'rf_f1': f1, 'rf_roc_auc': rf_roc_auc, 'lr_f1': lr_f1, 'lr_roc_auc': lr_roc_auc, 'rf_ub': rf_ub, 'rf_lb': rf_lb, 'lr_ub': lr_ub, 'lr_lb': lr_lb})
            
        print()
    return scores
```
